```python
!pip install pytesseract
!apt-get install tesseract-ocr
!apt-get install libtesseract-dev
!apt-get install tesseract-ocr-chi-sim
# 匯入所需的函式庫
import cv2
from google.colab import files
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt

# 上傳圖像
uploaded = files.upload()

# 讀取上傳的圖像
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# 將圖像轉換為灰度圖像
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, binary_image = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
denoised_image = cv2.GaussianBlur(binary_image, (5, 5), 0)
equalized_image = cv2.equalizeHist(denoised_image)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
dilated_image = cv2.dilate(equalized_image, kernel, iterations=1)
eroded_image = cv2.erode(dilated_image, kernel, iterations=1)

# 使用Tesseract進行車牌識別
custom_config = r'--oem 3 --psm 6'
plate_text = pytesseract.image_to_string(gray, config=custom_config)

# 顯示原始圖像和識別的車牌號碼
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("原始圖像")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.title("灰度圖像")
plt.axis("off")

plt.show()

# 印出識別結果
print("識別的車牌號碼:", plate_text.strip())
```

```python
!pip install pytesseract
!apt-get install tesseract-ocr
!apt-get install libtesseract-dev
!apt-get install tesseract-ocr-chi-sim

# 导入所需的库
import cv2
import pytesseract
from google.colab import files
import matplotlib.pyplot as plt

while True:
    # 显示文件上传按钮
    uploaded = files.upload()

    # 如果没有上传文件，退出循环
    if len(uploaded) == 0:
        break

    # 读取上传的图像
    image_path = list(uploaded.keys())[0]
    image = cv2.imread(image_path)

    # 将图像转换为灰度图像
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 使用Tesseract进行车牌识别
    custom_config = r'--oem 3 --psm 6'
    plate_text = pytesseract.image_to_string(gray, config=custom_config, lang='chi_sim')

    # 显示原始图像和识别的车牌号码
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("原始图像")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
    plt.title("灰度图像")
    plt.axis("off")

    plt.show()

    # 打印识别结果
    print("識別的車牌號碼:", plate_text.strip())
```
